```python
RomaniaMapFile = dict()
fileName = "/content/Test3.txt"
f1 = open(fileName, "r")
lines = f1.readlines()
for line in lines:
  if(len(line.split("\t")) != 2):
    continue
  parts = line.split("\t")
  key = parts[0].strip()
  neighbours = parts[1].strip().split(",")
  value = []
  for neighbour in neighbours:
    triple = (neighbour, None, 1)
    value.append(triple)
  RomaniaMapFile[key] = value

  sortedMap = dict(sorted(RomaniaMapFile.items(), key=lambda item: item[0], reverse=True))
  for key in sortedMap:
    neighbours = RomaniaMapFile[key]
    neighbourList = []
    for neighbour in neighbours:
      city,parent,weight = neighbour
      neighbourList.append(city)
    sortedList = sorted(neighbourList)
    sortedMap[key] = [(neighbour, None, 1) for neighbour in sortedList]

print(sortedMap)
# It stores the output of the above code in a .txt file
with open('20221CAI0087-Lab-02-output.txt', 'w') as f:
  f.write(str(sortedMap))
```
